```python
# let's again see through the dataset

import pandas as pd
import numpy as np

df = pd.read_csv('bangalore_tech_salaries.csv')
df.head()
```

```python
# let's rename some column names
# rename columns to snake_case

df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
)
df.columns       # so we made column names into lowercase and analysis suitable format
                # example :  Current CTC -> current_ctc   like this
```

```python
# standardize the role names now

df['role'].head(10)
```

```python
df['role'].tail(10)
```

```python
# standardize the company type now

df['company_type'].head(10)
```

```python
df['company_type'].tail(10)
```

```python
# standardize company name

company_mapping = {
    "early-stage" : "Early-stage",
    "EARLY-STAGE" : "Early-stage",
    "mid-size" : "Mid-size",
    "MID-SIZE" : "Mid-size",
    "Mid_size" : "Mid-size",
    "unicorn" : "Unicorn",
    "UNICORN" : "Unicorn",
    "mnc" : "MNC"
}

df["company_type_clean"] = df["company_type"].replace(company_mapping)
df["company_type_clean"]

df["company_type_clean"].value_counts()
```

```python
# standardize the education tier now

df['education_tier'].head(10)
```

```python
# standardize education tier name

education_mapping = {
    "2" : "Tier-2",
    "T2" : "Tier-2",
    "Tier 2" : "Tier-2",
    "1" : "Tier-1",
    "T1" : "Tier-1",
    "Tier 1" : "Tier-1",
    "3" : "Tier-3",
    "T3" : "Tier-3",
    "Tier 3" : "Tier-3"
}

df["education_tier_clean"] = df["education_tier"].replace(education_mapping)
df["education_tier_clean"]

df["education_tier_clean"].value_counts()
```

```python
# convert current CTC strings into Float (LPA)

df['current_ctc'].head(10)
```

```python
# convert CTC strings into Float (LPA)


df['current_ctc'] = (df['current_ctc'].str.strip()
.str.replace('LPA', '', regex = False)
.str.replace('lpa', '', regex = False)
.str.replace('Lpa', '', regex = False)
.str.replace('₹', '', regex = False)
.str.replace(',', '', regex = False)
.str.replace(' ', '', regex = False)
)


df['current_ctc'].head(10)
```

```python
# convert previous CTC strings into Float (LPA)


df['previous_ctc'] = (df['previous_ctc'].str.strip()
.str.replace('LPA', '', regex = False)
.str.replace('lpa', '', regex = False)
.str.replace('Lpa', '', regex = False)
.str.replace('₹', '', regex = False)
.str.replace(',', '', regex = False)
.str.replace(' ', '', regex = False)
)


df['previous_ctc'].head(10)
```

```python
# convert columns to numeric

df['current_ctc'] = pd.to_numeric(df['current_ctc'])
df['previous_ctc'] = pd.to_numeric(df['previous_ctc'])

# drop duplicates
df = df.drop_duplicates()
print(df.shape)
```

```python
# handle missing values
df['current_ctc'] = df['current_ctc'].fillna(df['current_ctc'].mean())
df['previous_ctc'] = df['previous_ctc'].fillna(df['previous_ctc'].mean())

df['role'] = df['role_clean'].fillna(df['role_clean'].mode()[0])
df['company_type'] = df['company_type_clean'].fillna(df['company_type_clean'].mode()[0])
df['education_tier'] = df['education_tier_clean'].fillna(df['education_tier_clean'].mode()[0])

df

print(df.dtypes)
```

```python
''' Q3.1 — CTC distribution by role. For every role, compute median, mean, min, and max CTC. Sort roles by
median descending. Which role pays the most? Which the least? Is the mean materially different from
the median in any role (sign of outliers)?

Approach :

I grouped the dataset by cleaned job role and calculated the mean, median, minimum and maximum current CTC for each role.
Then I sorted thre results by median salary in descending order to compare salary distributions across roles.

'''
```

```python
# let's see which role pays the least

print("Role with lowest median salary : ")
print(ctc_stats.tail(1))
```

```python
# check the outliers

ctc_stats["difference"] = ctc_stats["mean"] - ctc_stats["median"]
print(ctc_stats.sort_values(by="difference",ascending=False))
```

```python
''' Q3.2 — Experience curve for SDE Backend. Bin SDE Backend professionals by years_exp into 4 buckets
(0-1, 2-3, 4-5, 6+). Compute median CTC per bucket. What is the CTC growth per experience band? Hint:
pd.cut + groupby.

Approach :

I filtered the dataset for SDE Backend professionals, grouped them into four experience buckets using pd.cut()
and calculated the median current CTC for each bucket to analyze salary growth with experience.
```

```python
# Filter only SDE Backend professionals
backend_df = df[df["role_clean"] == "SDE Backend"].copy()

# Create experience buckets
bins = [0, 2, 4, 6, float("inf")]
labels = ["0-1", "2-3", "4-5", "6+"]

backend_df["experience_band"] = pd.cut(
    backend_df["years_exp"],
    bins=bins,
    labels=labels,
    right=False
)

# Calculate median CTC for each experience band
median_ctc = (
    backend_df.groupby("experience_band")["current_ctc"]
    .median()
)

print(median_ctc)
```

**Insight:**

1.   The median CTC for SDE Backend professionals increases consistently with experience, from 13.95 LPA (0–1 years) to 46.00 LPA (6+ years).
2.   The highest salary growth is observed for professionals with more than 6 years of experience, indicating that experience has a strong positive impact on compensation.


```python
''' Q3.3 — Skill premium for SDEs. Pick 4 skills from {AWS, ML, System Design, Kubernetes, Deep Learning,
Microservices}. For each, compare median CTC of SDEs (any of Backend / Frontend / Full-Stack) who
have the skill vs those who don't. Which skill adds the most to median CTC? Hint: .str.contains on the
cleaned skills column

Approach :

I filtered the dataset to include only SDE professionals (Backend, Frontend, and Full-Stack).
Then, for each selected skill, I divided professionals into two groups—those who have the skill and those who do not.
Finally, I calculated the median current CTC for both groups
and compared the salary difference to identify the skill with the highest salary premium. '''
```

```python
df.columns
```

```python
# Filter only SDE professionals
sde_df = df[
    df["role_clean"].str.contains(
        "Backend|Frontend|Full.?Stack",
        case=False,
        na=False
    )
].copy()

skills = ["AWS", "ML", "System Design", "Deep Learning"]

results = []

for skill in skills:

    has_skill = sde_df[
        sde_df["skills"].str.contains(skill, case=False, na=False)
    ]["current_ctc"].median()

    no_skill = sde_df[
        ~sde_df["skills"].str.contains(skill, case=False, na=False)
    ]["current_ctc"].median()

    premium = has_skill - no_skill

    results.append([skill, has_skill, no_skill, premium])

result_df = pd.DataFrame(
    results,
    columns=[
        "Skill",
        "Median CTC (Has Skill)",
        "Median CTC (No Skill)",
        "Premium"
    ]
)

result_df = result_df.sort_values(by="Premium", ascending=False)

print(result_df)
```

```python
print("\nSkill with Highest Premium:")
print(result_df.iloc[0])
```

**Insight:**



1.   Among the selected skills, System Design provides the highest median CTC premium of 2.80 LPA for SDE professionals, followed by ML (1.85 LPA) and AWS (1.60 LPA).
2.   Deep Learning shows a slightly negative premium (-0.40 LPA), indicating no salary advantage for this dataset.



```python
''' Q3.4 — Company-type premium, same role. Pick one role (SDE Backend or Data Scientist). Compute
median CTC across Unicorn / MNC / Mid-size / Early-stage for that role. What is the percentage
premium of Unicorn over MNC at the same role?

Approach :

I selected the Data Scientist role and grouped the data by company type.
Then I calculated the median current CTC for each company type
and computed the percentage salary premium of Unicorn companies over MNCs.
'''
```

```python
# Filter Data Scientist role
role_df = df[df["role_clean"] == "Data Scientist"]

# Median CTC by company type
median_ctc = (
    role_df.groupby("company_type_clean")["current_ctc"]
    .median()
)

print(median_ctc)
```

```python
# calculate unicorn premium over MNC

unicorn = median_ctc["Unicorn"]
mnc = median_ctc["MNC"]

premium = ((unicorn - mnc) / mnc) * 100

print(f"Unicorn Premium over MNC: {premium:.2f}%")
```

**Insight:**



1.   Data Scientists working in Unicorn companies have the highest median CTC.
2.   Their median salary is 86.44% higher than that of Data Scientists working in MNCs, showing a clear salary premium for Unicorn companies.


```python
''' Q3.5 — Underpaid professionals (multi-step). For every (role_clean, company_type_clean, exp_band)
combination, compute median CTC. Then for every individual, compute the gap = their_ctc -
their_group_median. Print the top 10 most-underpaid professionals with name, role, company_type,
exp, ctc, group_median, gap. Filter to groups with at least 10 members (so the median is reliable). Hint:
groupby + transform + filter.

Approach :

I grouped employees based on role, company type, and experience band.
I calculated the median CTC for each group and assigned it to every employee.
Then I computed the salary gap (Current CTC − Group Median), filtered groups with at least 10 employees
and displayed the top 10 employees with the largest negative salary gap.

'''
```

```python
# Create experience bands

bins = [0, 2, 4, 6, float("inf")]
labels = ["0-1", "2-3", "4-5", "6+"]

df["exp_band"] = pd.cut(
    df["years_exp"],
    bins=bins,
    labels=labels,
    right=False
)
```

```python
# compute group median

group_cols = ["role_clean", "company_type_clean", "exp_band"]

df["group_median"] = (
    df.groupby(group_cols)["current_ctc"]
      .transform("median")
)
```

```python
# count members in each group

df["group_size"] = (
    df.groupby(group_cols)["employee_id"]
      .transform("count")
)

# compute salary gap
df["gap"] = df["current_ctc"] - df["group_median"]

# filter reliable groups
filtered_df = df[df["group_size"] >= 10]

# AI-assisted: Used ChatGPT to understand grouping and aggregation.
```

```python
# Top 10 underpaid employees

underpaid = (
    filtered_df.sort_values("gap")
    .head(10)
)

print(
    underpaid[
        [
            "employee_id",
            "role_clean",
            "company_type_clean",
            "exp_band",
            "current_ctc",
            "group_median",
            "gap"
        ]
    ]
)
```

**Insight:**

1.   The analysis identifies employees whose current CTC is below the median salary of professionals with the same role, company type, and experience band.
2.   The most underpaid employee is BLR0887 (Data Scientist, MNC, 2–3 years), earning 19.7 LPA, which is 6.9 LPA below the group median of 26.6 LPA.
